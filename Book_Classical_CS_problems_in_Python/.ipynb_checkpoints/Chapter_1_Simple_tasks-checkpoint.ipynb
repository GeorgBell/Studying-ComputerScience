## 1.1. Fibonacci
- Recursive algorithm

```python
def fib_recursive(num):
    if num < 2:
        return num
    return fib_recursive(num - 1) + fib_recursive(num - 2)
```

Let's check it out

```python
for i in range(10):
    print(fib_recursive(i))
```

- Adding recursion tracing to the algorithm

```python
num_calls = 0

def fib_recursive_modif(num, level=0):
    global num_calls
    num_calls += 2
    level += 1
    print(f"Num: {num}, level: {level}")
    if num < 2:
        return num
    return fib_recursive_modif(num - 1, level) + fib_recursive_modif(num - 2, level)

print(f"Fib value: {fib_recursive_modif(20)}")
print(f"Number of calls: {num_calls}")
```

- Memorizing with dictionary

```python
num_calls = 0
dict_memo = {0: 0, 1:1}

def fib_recursive_dictmemo(num, level=0):
    level += 1
    global num_calls
    num_calls += 1
    print(f"Num: {num}, level: {level}")
    if num not in dict_memo:
        dict_memo[num] = fib_recursive_dictmemo(num-1, level) + fib_recursive_dictmemo(num-2, level)
    return dict_memo[num]

print(fib_recursive_dictmemo(20))
print(num_calls)
```

- Memorizing with decorator

```python
from functools import lru_cache
num_calls = 0

@lru_cache(maxsize=None)
def fib_recursive_modif(num, level=0):
    global num_calls
    num_calls += 2
    level += 1
    print(f"Num: {num}, level: {level}")
    if num < 2:
        return num
    return fib_recursive_modif(num - 1, level) + fib_recursive_modif(num - 2, level)

print(f"Fib value: {fib_recursive_modif(20)}")
print(f"Number of calls: {num_calls}")
```

- Iterative implementation

```python
def fib_iterative(num: int) -> int:
    prev: int = 0
    cur: int = 1
    
    for _ in range(1, num):
        prev, cur = cur, prev + cur
    return cur
    
print(fib_iterative(20))
```

- Yielding

```python
from typing import Generator

def fib_generator(num: int) -> Generator[int, None, None]:
    yield 0
    if num > 0:
        yield 1
        
    prev: int = 0
    cur: int = 1
    
    for _ in range(1, num):
        prev, cur = cur, prev + cur
        yield cur

for i in fib_generator(20):
    print(i)
```

## 1.2. Simple compression
Gene example: A, C, G, T: from chars to bits

```python
class CompressedGene:
    def __init__(self, gene: str) -> None:
        self._compress(gene)
    
    def _compress(self, gene: str) -> None:
        self.bit_string: int = 1 # Starting bit
            
        for nucleotide in gene.upper():
            self.bit_string <<= 2
            print(self.bit_string)
            print(nucleotide)
            
            if nucleotide == "A":
                self.bit_string |= 0b00
            elif nucleotide == "C":
                self.bit_string |= 0b01
            elif nucleotide == "G":
                self.bit_string |= 0b10
            elif nucleotide == "T":
                self.bit_string |= 0b11
            else:
                raise ValueError(f"Invalid Nucleotide: {nucleotide}")
    
    def decompress(self) -> str:
        gene: str = ""
        for i in range(0, self.bit_string.bit_length() - 1, 2):
            bits: int = self.bit_string >> i & 0b11
            if bits == 0b00:
                gene += "A"
            elif bits == 0b01:
                gene += "C"
            elif bits == 0b10:
                gene += "G"
            elif bits == 0b11:
                gene += "T"
            else:
                raise ValueError(f"Invalid bits: {bits}")
        return gene[::-1]
    
    def __str__(self) -> str:
        return self.decompress()
```

```python
from sys import getsizeof

gene_seq: str = "ACTGGATG"
print(gene_seq)
print(f"Original is {getsizeof(gene_seq)} bytes")

compressed: CompressedGene = CompressedGene(gene_seq)
print(f"Compressed is {getsizeof(compressed)} bytes")

print(compressed)
```

## 1.3. Ciphering
Initial data XOR Fiction data (key 1) = Result (key 2)

```python
from secrets import token_bytes
from typing import Tuple

def random_key(length: int) -> int:
    tb: bytes = token_bytes(length)
    print(tb)
    return int.from_bytes(tb, "big")
```

```python
random_key(1)
```

```python
int.from_bytes(b'\xe0', "big")
```

```python
def encrypt(original: str) -> Tuple[int, int]:
    key_dummy: int = random_key(len(original))
    # str => bytes => int
    original_bytes: bytes = original.encode()
    original_int: int = int.from_bytes(original_bytes, "big")
    # XOR
    encrypted: int = original_int ^ key_dummy
    return encrypted, key_dummy

original_str = "Hello World!"
encrypted_str, key = encrypt(original_str)
print(encrypted_str)
print(key)
```

```python
print(f"Size of original str: {getsizeof(original_str)}")
print(f"Size of encrypted str: {getsizeof(encrypted_str)}")
```

```python
def decrypt(encrypted_str: int, key: int) -> str:
    decrypted: int = encrypted_str ^ key
    decrypted_bytes: bytes = decrypted.to_bytes((decrypted.bit_length() + 7) // 8, "big")
    return decrypted_bytes.decode()

decrypt(encrypted_str, key)
```

## 1.4. Computing Pi value
Using Leibnitz formula

```python
def calculate_pi(num_signs: int) -> float:
    nominator: float = 4.0
    denominator: float = 1.0
    sign: float = 1.0
    pi: float = 0.0
    for _ in range(num_signs+1):
        pi += nominator / denominator * sign
        denominator += 2.0
        sign *= -1.0
    return pi

calculate_pi(1000000)
```

## 1.5. Hanoi tower

```python
from typing import TypeVar, Generic, List
T = TypeVar('T') # Not specified type

class Stack(Generic[T]):
    def __init__(self) -> None:
        self._container: List[T] = []
    
    def push(self, item: T) -> None:
        self._container.append(item)
    
    def pop(self) -> T:
        return self._container.pop()
    
    def __repr__(self) -> str:
        return repr(self._container)
```

```python
my_stack = Stack()
my_stack.push(1)
my_stack.push(10)
```

- Let's build 3 hanoi towers with 3 disks on the 1st

```python
def create_task(num_disks:int) -> Tuple:
    tower_a: Stack = Stack()
    tower_b: Stack = Stack()
    tower_c: Stack = Stack()

    for disc in range(1, num_disks+1):
        tower_a.push(disc)
    return tower_a, tower_b, tower_c
```

- Let's solve hanoi towers problem with recursive algorithm

```python
def hanoi_solution(begin: Stack[int], end: Stack[int], temp: Stack[int], n: int, level: int = 0) -> None:
    level += 1
    if n == 1:
        global num_moves
        num_moves += 1
        print("before pushing")
        print(level, begin, end, temp)
        end.push(begin.pop())
        print("After pushing")
        print(level, begin, end, temp)
        
    else:
        hanoi_solution(begin, temp, end, n-1, level)
        hanoi_solution(begin, end, temp, 1, level)
        hanoi_solution(temp, end, begin, n-1, level)
```

```python
num_moves = 0
num_disks = 3
tower_a, tower_b, tower_c = create_task(num_disks)

print("Before:")
print(f"Tower A: {tower_a} \nTower B: {tower_b} \nTower C: {tower_c}")

hanoi_solution(tower_a, tower_c, tower_b, num_disks)
print("After:")
print(f"Tower A: {tower_a} \nTower B: {tower_b} \nTower C: {tower_c}")
print(f"Number of calls: {num_calls}")
```
